```python
import matplotlib.pyplot as plt

# Dados
dados = sorted([4, 5, 6, 6.5, 7, 8, 8.5, 9, 10])
n = len(dados)

# --- Cálculo manual dos quartis (sem arredondamento) ---
def calcular_quartil(dados, percent):
    pos = (n + 1) * percent / 100
    base = int(pos)
    frac = pos - base
    if base == 0:
        return dados[0]
    elif base >= n:
        return dados[-1]
    else:
        return dados[base - 1] + frac * (dados[base] - dados[base - 1])

q1 = calcular_quartil(dados, 25)
q2 = calcular_quartil(dados, 50)
q3 = calcular_quartil(dados, 75)
minimo = min(dados)
maximo = max(dados)

# Exibir valores exatos no console
print(f"Mínimo: {minimo}")
print(f"Q1: {q1}")
print(f"Mediana (Q2): {q2}")
print(f"Q3: {q3}")
print(f"Máximo: {maximo}")

# --- Gráfico Boxplot na vertical ---
fig, ax = plt.subplots()
ax.boxplot(dados, vert=True, patch_artist=True)

# Título e rótulo
plt.title('Boxplot - Distribuição dos Dados (Quartis Exatos)')
plt.ylabel('Valores')

# Adicionar os textos com os valores exatos
plt.text(1.05, minimo, f'Mín = {minimo}', va='center', color='purple')
plt.text(1.05, q1, f'Q1 = {q1}', va='center', color='blue')
plt.text(1.05, q2, f'Q2 = {q2}', va='center', color='green')
plt.text(1.05, q3, f'Q3 = {q3}', va='center', color='red')
plt.text(1.05, maximo, f'Máx = {maximo}', va='center', color='purple')

# Ajuste do eixo y
plt.ylim(minimo - 1, maximo + 1)

# Mostrar gráfico
plt.show()


```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# === Dados base ===
dados = {
    'Aluno': [f'Aluno {i}' for i in range(1, 16)],
    'Tempo de Estudo (h)': [1, 2, 3, 3, 4, 5, 5, 6, 6, 7, 8, 8, 9, 9, 10],
    'Nota': [4, 5, 5, 6, 6, 6, 7, 7, 8, 8, 8, 9, 9, 9, 10]
}

df = pd.DataFrame(dados)

# === Cálculo das correlações ===
pearson_corr = df['Tempo de Estudo (h)'].corr(df['Nota'], method='pearson')
spearman_corr = df['Tempo de Estudo (h)'].corr(df['Nota'], method='spearman')
kendall_corr = df['Tempo de Estudo (h)'].corr(df['Nota'], method='kendall')

print("Coeficientes de Correlação:")
print(f"→ Pearson:  {pearson_corr:.3f}")
print(f"→ Spearman: {spearman_corr:.3f}")
print(f"→ Kendall:  {kendall_corr:.3f}")

# === Gráfico com reta e labels ===
x = df['Tempo de Estudo (h)']
y = df['Nota']

# Regressão linear (para visualizar Pearson)
m, b = np.polyfit(x, y, 1)

plt.figure(figsize=(8,6))
plt.scatter(x, y, color='blue')

# Labels
for i, aluno in enumerate(df['Aluno']):
    plt.text(x[i]+0.1, y[i]+0.05, aluno, fontsize=8)

# Reta
plt.plot(x, m*x + b, color='red', linewidth=2, label='Reta de Tendência (Pearson)')

plt.title('Correlação entre Tempo de Estudo e Nota')
plt.xlabel('Tempo de Estudo (h)')
plt.ylabel('Nota')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression

np.random.seed(42)

# === Geração dos dados ===
n = 50
tempo_estudo = np.random.uniform(1, 10, n)
nota = 3 + 0.6 * tempo_estudo + np.random.normal(0, 1, n)

df = pd.DataFrame({'Tempo de Estudo (h)': tempo_estudo, 'Nota': nota})

# === Regressão linear ===
X = df[['Tempo de Estudo (h)']]
y = df['Nota']
modelo = LinearRegression()
modelo.fit(X, y)

b0 = modelo.intercept_
b1 = modelo.coef_[0]
print(f"Equação: Nota = {b0:.2f} + {b1:.2f} * Tempo de Estudo")

# === Plot ===
plt.figure(figsize=(8, 6))
sns.regplot(x='Tempo de Estudo (h)', y='Nota', data=df, color='blue', line_kws={'color': 'red'})

# Ponto especial para ilustrar o erro
x_exemplo = 6
y_real = 3 + 0.6 * x_exemplo + np.random.normal(0, 1, 1)[0]  # simula um ponto real
y_prev = b0 + b1 * x_exemplo  # valor previsto pela reta

# Linha vertical do erro (resíduo)
plt.plot([x_exemplo, x_exemplo], [y_prev, y_real], 'k--', lw=2)
plt.scatter(x_exemplo, y_real, color='green', s=80, label='Valor real (ponto observado)')
plt.scatter(x_exemplo, y_prev, color='red', s=80, label='Valor previsto pela reta')

# === Labels explicativos ===
plt.text(1, b0 + 0.2, 'β₀ (Intercepto)\nNota quando tempo de estudo = 0', fontsize=10, color='black')
plt.text(8, b0 + b1*8 + 0.3, 'β₁ (Inclinação)\nVariação da nota a cada hora estudada', fontsize=10, color='red')
plt.text(x_exemplo + 0.3, (y_prev + y_real)/2, 'ε (Erro)', fontsize=10, color='purple')

# Título e eixos
plt.title(f"Regressão Linear Simples: Nota = {b0:.2f} + {b1:.2f} × Tempo de Estudo", fontsize=12)
plt.xlabel('Tempo de Estudo (h)')
plt.ylabel('Nota')
plt.legend()
plt.grid(True)
plt.show()

```

```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Dados
dados = pd.DataFrame({
    'horas_estudo': [1, 2, 3, 4, 5, 6, 7],
    'frequencia': [60, 65, 70, 75, 80, 85, 90],
    'nota_prova': [50, 55, 60, 68, 73, 78, 85]
})

# Define variáveis independentes (X) e dependente (Y)
X = dados[['horas_estudo', 'frequencia']]
Y = dados['nota_prova']

# Adiciona constante (intercepto)
X = sm.add_constant(X)

# Cria e ajusta o modelo
modelo_multiplo = sm.OLS(Y, X).fit()

# Exibe o resumo estatístico
print(modelo_multiplo.summary())

# ====================================
# Gráfico didático (2D)
# ====================================

# Gera os valores previstos pelo modelo
dados['nota_prevista'] = modelo_multiplo.fittedvalues

# Cria o gráfico
plt.figure(figsize=(8, 6))
sns.scatterplot(x='horas_estudo', y='nota_prova', data=dados, color='blue', label='Notas reais')
sns.lineplot(x='horas_estudo', y='nota_prevista', data=dados, color='red', label='Reta prevista')

plt.title('Regressão Linear Múltipla (Nota ~ Horas de Estudo + Frequência)')
plt.xlabel('Horas de Estudo')
plt.ylabel('Nota na Prova')
plt.legend()
plt.grid(True)
plt.show()

# Exibe também a diferença entre real e previsto
print("\nComparação entre notas reais e previstas:")
print(dados[['horas_estudo', 'frequencia', 'nota_prova', 'nota_prevista']])
```

```python
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns

# Dados
dados = pd.DataFrame({
    'horas_estudo': [1, 2, 3, 4, 5, 6, 7],
    'frequencia': [60, 65, 70, 75, 80, 85, 90],
    'nota_prova': [50, 55, 60, 68, 73, 78, 85]
})

# Define variáveis independentes (X) e dependente (Y)
X = dados[['horas_estudo', 'frequencia']]
Y = dados['nota_prova']

# Adiciona constante (intercepto)
X = sm.add_constant(X)

# Cria e ajusta o modelo
modelo_multiplo = sm.OLS(Y, X).fit()

# Adiciona valores previstos
dados['nota_prevista'] = modelo_multiplo.fittedvalues

# ==============================
# Gráfico com frequência em cor
# ==============================
plt.figure(figsize=(8, 6))
scatter = plt.scatter(
    x=dados['horas_estudo'],
    y=dados['nota_prova'],
    c=dados['frequencia'],
    cmap='viridis',
    s=100,
    edgecolor='k'
)

# Linha das previsões
sns.lineplot(
    x='horas_estudo',
    y='nota_prevista',
    data=dados,
    color='red',
    label='Linha prevista'
)

plt.title('Regressão Linear Múltipla (Nota ~ Horas de Estudo + Frequência)')
plt.xlabel('Horas de Estudo')
plt.ylabel('Nota na Prova')
plt.colorbar(scatter, label='Frequência (%)')
plt.legend()
plt.grid(True)
plt.show()

# Comparação real x previsto
print("\nComparação entre notas reais e previstas:")
print(dados[['horas_estudo', 'frequencia', 'nota_prova', 'nota_prevista']])
```
